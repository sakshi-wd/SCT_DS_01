```python
# Task 01: Age Distribution Bar Chart

This notebook visualizes India's population distribution by age group using a bar chart. It demonstrates how to work with a sample dataset and generate visual insights using Python and Matplotlib.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df_population = pd.read_csv("age_distribution_dataset.csv")
df_population
```

```python
# Plot a bar chart
plt.figure(figsize=(8, 5))
plt.bar(df_population['Age_Group'], df_population['Population (millions)'], color='pink')
plt.xlabel('Age Group')
plt.ylabel('Population (millions)')
plt.title("India's Population Distribution by Age (2022)")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig("age_distribution_chart.png")
plt.show()
```
